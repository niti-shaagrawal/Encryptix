```python
# undersampling the data because the data is highly unbalamced

legit_sample= legit.sample(n=473)
```

```python
# conacting both dataframes to build a more balanced dataset 

new_data= pd.concat((legit_sample, fraud), axis=0)
```

```python
# comparing the dataset values on the basis of their class to show that the baisc distribution of the dataset is not affected by undersampling 

new_data.groupby('Class').mean()
```

```python
# dropping the exra columns 

x= new_data.drop(columns=['Class'])
y= new_data['Class']
```

```python
# spliiting the data into training and testing data 

x_train, x_test, y_train, y_test= train_test_split(x,y, test_size= 0.2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
print(y.shape, y_train.shape, y_test.shape)
```

```python
# building the model 

model= LogisticRegression(max_iter=10000)
```

```python
model.fit(x_train, y_train)
```

```python
# model evaluation and accuracy score

x_train_pred= model.predict(x_train)
print(x_train_pred)
```

```python
# accuracy score for training data 

train_accuracy= accuracy_score(x_train_pred,y_train)
print('The accuracy score for the training data is: ',  train_accuracy)
```

```python
# accuracy score for testing data 

x_test_pred= model.predict(x_test)
print(x_test_pred)
```

```python
test_accuracy= accuracy_score(x_test_pred, y_test)
print('The accuracy score for the testing data is: ', test_accuracy)
```

```python
# precision score for training and testing data 

train_precision = precision_score(x_train_pred, y_train)
print('The precision score for training data is: ', train_precision)
test_precision= precision_score(x_test_pred, y_test)
print('The precision score for testing data is: ', test_precision)
```

```python
# f1 score for training and testing data 

train_f1 = f1_score(x_train_pred, y_train)
print('The f1 score for training data is: ', train_f1)
test_f1= f1_score(x_test_pred, y_test)
print('The f1 score for testing data is: ', test_f1)
```

```python
# recall_score for training and testing data 

train_recall = recall_score(x_train_pred, y_train)
print('The recall score for training data is: ', train_recall)
test_recall= recall_score(x_test_pred, y_test)
print('The recall score for testing data is: ', test_recall)
```
